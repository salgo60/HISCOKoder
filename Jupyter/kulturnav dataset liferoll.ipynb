```python
import urllib3, json
import pandas as pd 
http = urllib3.PoolManager() 
pd.set_option("display.max.columns", None) 

url = "https://kulturnav.org/api/list/23c7080c-6752-4691-8c8a-ce0d65623c51"
r = http.request('GET', url)
data = json.loads(r.data)
dftot = pd.DataFrame(data)
```

```python
dftot.head(10)
```

```python
dftot.info()
```

```python
# Flattening
dfN = pd.json_normalize(data)
dfN.head(5)
```

```python
dfN.info()
```

```python
dfN["entityTypeName.sv"].value_counts()
```

```python
dfNsort = dfN[{"caption.sv","entityTypeName.sv"}].sort_values(by=["entityTypeName.sv","caption.sv"])
dfNsort
```

```python
dfNsort.to_csv("kulturnav_liferoll.csv")
```
