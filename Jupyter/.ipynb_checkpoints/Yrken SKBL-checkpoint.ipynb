```python
from datetime import datetime
start_time  = datetime.now()
print("Last run: ", start_time)
```

```python
import requests
import urllib.request
import time
from bs4 import BeautifulSoup
def cleanupstring(s):
    #remove span tag
    try:
        soupStr = BeautifulSoup(s, "html.parser")
        a_tag = soupStr.a
        span_tag = soupStr.span.extract()
    except Exception as e:
            #print("\tError",s, e )
            return
    return a_tag

urls = [
    'https://skbl.se/sv/nyckelord']

newList = []

for url in urls:

    response = requests.get(url)
#    soup = BeautifulSoup(response.text.encode('UTF-8'), "html.parser")
    soup = BeautifulSoup(response.text, "html.parser")
    soup.findAll('a')
    one_a_tag = soup.findAll('a')[0]
    #   link = one_a_tag['href']
    time.sleep(1)
    for i in range(1, len(soup.findAll('a'))):
        new_item = dict()
        one_a_tag = soup.findAll('a')[i]
        try:
            one_a_tag_href = one_a_tag.get('href')
            #print("\tone_a_tag_href\t",one_a_tag_href)
            if "nyckelord" in one_a_tag_href:
                cleanedTag = cleanupstring(one_a_tag.prettify()) 
                #print(i, "\t", cleanedTag,"\t",one_a_tag)
                #if not "TOPPEN" in one_a_tag_href.text:
                newList.append(cleanedTag.text)
        except Exception as e:
            #print("error", e )
            pass
print (len(newList) ," antal poster")
```

```python
import pandas as pd  
newListdf = pd.DataFrame(newList,columns=list(["Yrke"]))
newListdf["Yrke"] = newListdf["Yrke"].str.strip()
#newListdf.info()
newListdf
```
